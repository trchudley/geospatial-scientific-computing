## What's wrong with standard python for maths?

Standard python does give us some mathematical capability, primarily through the `math` module:

```python
import math

# show some simple examples of what you can do here
```

The main problem is that the standard data types available to us aren't naturally good at doing large mathematical operations. Let's, for example, imaging we have some temperatures in celcius (in a `list` form), and want to convert them to kelvin:

```python
temperatures_c = [18.2, 19.1, 20.5, 21.0]
```

```python
temperatures_k = temperatures_c + 273.15
```

Uh oh. Python has thrown an error! It doesn't care what's inside the list (be they strings, integers, dictionaries, etc...), so it doesn't know what we mean when we try and "add one to a list".

Okay, let's take another tack. Let's create a list of equal length, and try and add each element together. This makes sense, right?

```python
add_kelvin = [273.15, 273.15, 273.15, 273.15]

temperatures_k = temperatures_c + add_kelvin

temperatures_k
```

This still isn't what we want! Python thinks that "adding" lists together means _concatenating_ lists, because lists are designed to handle all types of objects, not just numbers.

In fact, the easiest way to perform this is probably to loop through every temperature individually, adding values:

```python
temperatures_k = [t + 273.15 for t in temperatures_c]

temperatures_k
```

This isn't convenient at all. We need a tool that allows us to operate more easily on large numerical datasets.

## `numpy`

```python
import numpy as np
```

`numpy`, conventionally shortened to `np` when importing, introduces another type of object: the numpy array.

```python
temperatures_c = np.array([18.2, 19.1, 20.5, 21.0])

temperatures_c
```

Because we know that everything inside an array is numerical, our instinctive operations from before work:

```python
temperatures_c + 274.15
```

```python
add_kelvin = np.array([273.15, 273.15, 273.15, 273.15])

temperatures_c + add_kelvin
```
