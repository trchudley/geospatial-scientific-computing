# A Python and Jupyter Primer

The rest of this week aims to give you a very brief overview of the Python syntax, designed for those with little prior experience in Python. It will form the foundations of how we interact with Python for the rest of the course. If you familiar with Python, then most of this will not surprise you, although it is worth going through once to make sure. The important thing is that we have spent our lab time this week to ensure we are comfortable with installing Python and loading Jupyter Notebooks. 

However, if you are not relatively new to Python, then it is critical you take the time to understand the basics of what's going on, and make sure you're comfortable reading through and running Python code.

## Opening a new notebook

Open up your command prompt/terminal and run Jupyter Lab (run `conda activate gsc`; and then type `jupyter lab`).

If you are happy you are in the right directory (navigate around using the left-hand sidebar), then you can create a new notebook by clicking a 'new notebook' button in the middle of the page.
